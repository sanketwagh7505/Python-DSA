### 14. Write a program that will determine weather when the value of temperature and humidity is provided by the user

```python
temperature = float(input("Enter the Temperature in Celcius: "))
humidity = float(input("Enter Humidity: "))

if temperature>= 30 and humidity >=90:
    print("Hot and Humid")
elif temperature>=30 and humidity<90:
    print("Hot")
elif temperature<30 and humidity >=90:
    print("Cool and Humid")
else:
    print("Cool")            
```
